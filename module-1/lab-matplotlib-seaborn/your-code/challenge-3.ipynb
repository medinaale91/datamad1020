#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
plt.figure(figsize=(15,5))

plt.subplot(1,3,1) # (nrows, ncols, index)
sns.histplot(titanic.Age, bins = 10)
plt.title("Titanic Ages bins = 10");
plt.xlabel("Age");
plt.grid(alpha = 0.5); #or plt.grid(axis="y", alpha = 0.5) to show only the grid in one axis.

plt.subplot(1,3,2)
sns.histplot(titanic.Age, bins = 20)

plt.subplot(1,3,3)
sns.histplot(titanic["Age"], bins = 50);
```

#### Whether the bin size affects your plot? Why?

Yes it does, as it gathers ages depending on the amount of bins we want to split the histogram. With 50 Bins, the information is more precise as it allows you to split into smaller sections, showing more detail.

Si, cuanto mayor sea el nmero de bins, mayor detalle de la muestra y ms precisa es la informacion. La primera tabla muestra las edades divididas en 10 bins, mientras que la ultima tabla muestra las edades repartidas 
entre 50 bins, por lo que el rango de edades es mucho menor y se permite conocer con ms exactitud los datos.
