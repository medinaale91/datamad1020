#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)

plt.subplot(1,2,1) #(nrows, ncols, index)
plt.title("Graphic 1")
plt.plot(x,y, "g*")

plt.subplot(1,2,2)
plt.title("Graphic 2")
plt.plot(x,z, "r--")
```

```python
# your code here-2st way (call `subplots` only once not using the `index` parameter)
fig,a = plt.subplots(1,2)
a[0].plot(x,y,"b*")
a[1].plot(x,z,"y--")
a[0].set_title("Graphic 1")
a[1].set_title("Graphic 2")
plt.show()
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
# your code here
fig,a = plt.subplots(1,2, figsize = (13,5))
a[0].plot(x, x**2, x, np.exp(x),"g")
a[1].plot(x, x**2, x, np.exp(x),"r")

#a[0].plot(x,x**2,"g")
#a[0].plot(x,np.exp(x),"g")

#a[1].plot(x,x**2,"r")
#a[1].plot(x,np.exp(x),"r")

a[0].set_xlabel("X")
a[0].set_ylabel("Y")
a[1].set_xlabel("X")
a[1].set_ylabel("Y")

a[0].set_title("Normal scale")
a[1].set_title("Logarithmic scale")

a[1].set_yscale("log")
plt.show()
```
