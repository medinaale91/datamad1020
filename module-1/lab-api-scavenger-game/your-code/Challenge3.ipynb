```python
from dotenv import load_dotenv
import os
import re
import pandas as pd
import requests
```

```python
load_dotenv()
```

```python
gh_token = os.getenv("GITHUB_TOKEN")
gh_token
```

```python
url = "https://api.github.com"
headers = {
    "Authorization": f"token {gh_token}"
}
```

```python
endpoint = "/repos/{owner}/{repo}/contents"
repo_info = {
    "owner":"ironhack-datalabs",
    "repo":"scavenger"
}
```

```python
data = requests.get(url+endpoint.format(**repo_info), 
                    headers=headers)
```

```python
data = data.json()
```

```python
data
```

```python
content = [content["url"] for content in data if content["type"] == "dir"]
content
```

```python
raw_urls = []
for url in content:
    data = requests.get(url)
    json_array = data.json()
    for dictionary in json_array:
        if "scavengerhunt" in dictionary["name"]:
            raw_urls.append(dictionary["download_url"])
        
print(raw_urls)
```

```python
def by_last_path(url):
    return url[-18:]

sorted_raw_urls = sorted(raw_urls, key = by_last_path)
print(sorted_raw_urls)
```

```python
cold_joke = []
for raw_urls in sorted_raw_urls:
    data = requests.get(raw_urls)
    clean_text = data.text.strip()
    cold_joke.append(clean_text)
    
print(" ".join(cold_joke))
```
